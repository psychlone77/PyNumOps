# Fixed Point Iteration Method

This notebook demonstrates the Fixed Point Iteration Method for finding the roots of a function.

## Introduction

The Fixed Point Iteration Method is an iterative method for solving equations of the form $x = g(x)$. The method starts with an initial guess $x_0$ and iteratively applies the function $g(x)$ to get closer to the root.

### Algorithm
1. Choose an initial guess $x_0$.
2. Compute the next approximation using $x_{n+1} = g(x_n)$.
3. Repeat step 2 until the difference between $x_{n+1}$ and $x_n$ is less than a specified tolerance.

### Example
Consider the equation $x = \cos(x)$. We can rewrite this as $g(x) = \cos(x)$ and use the Fixed Point Iteration Method to find the root.

## Mathematical Solution

To solve the equation $x = \cos(x)$ using the Fixed Point Iteration Method, follow these steps:

1. **Choose an initial guess**: Let's start with $x_0 = 0.5$.

2. **Iterate using the function $g(x) = \cos(x)$**:
   $$
   \begin{align*}
   x_1 &= g(x_0) = \cos(0.5) \approx 0.8776 \\
   x_2 &= g(x_1) = \cos(0.8776) \approx 0.6390 \\
   x_3 &= g(x_2) = \cos(0.6390) \approx 0.8027 \\
   x_4 &= g(x_3) = \cos(0.8027) \approx 0.6948 \\
   x_5 &= g(x_4) = \cos(0.6948) \approx 0.7682 \\
   \end{align*}
   $$

3. **Continue iterating until convergence**:
   $$
   \begin{align*}
   x_6 &= g(x_5) = \cos(0.7682) \approx 0.7192 \\
   x_7 &= g(x_6) = \cos(0.7192) \approx 0.7524 \\
   x_8 &= g(x_7) = \cos(0.7524) \approx 0.7300 \\
   x_9 &= g(x_8) = \cos(0.7300) \approx 0.7451 \\
   x_{10} &= g(x_9) = \cos(0.7451) \approx 0.7356 \\
   \end{align*}
   $$

4. **Check for convergence**: The iterations continue until the difference between successive approximations is less than a specified tolerance. For example, if we set the tolerance to $10^{-5}$, we stop when $|x_{n+1} - x_n| < 10^{-5}$.

## Example 2:

Find the first approximate root of the equation $\cos(x) = 3x - 1$ up to 4 decimal places.

### Solution:

Let $f(x) = \cos(x) - 3x + 1 = 0$

As per the algorithm, we find the value of $x_0$, for which we have to find $a$ and $b$ such that $f(a) < 0$ and $f(b) > 0$.

Now,
$$
f(0) = 2
$$

$$
f\left(\frac{\pi}{2}\right) = -3\left(\frac{\pi}{2}\right) - 1 < 0
$$

Hence, $x_0$ is a value lying between $0$ and $\frac{\pi}{2}$, for ease of calculation let us take $x_0 = 0$.

Now, we shall find $g(x)$ such that $|g'(x)| < 1$ at $x = x_0$.

$$
\cos(x) - 3x + 1 = 0
$$

$$
\Rightarrow x = \frac{\cos(x) + 1}{3}
$$

Then $g(x) = \frac{\cos(x) + 1}{3}$ which satisfies $|g'(x)| < 1$ at $x = 0$.

Now, applying the iterative method $x_n = g(x_{n-1})$ for $n = 1, 2, 3, 4, 5, \ldots$

For $n = 1$:
$$
x_1 = g(x_0) = \frac{\cos(0) + 1}{3} = \frac{2}{3} = 0.66667
$$

For $n = 2$:
$$
x_2 = g(x_1) = \frac{\cos(0.66667) + 1}{3} = 0.595296
$$

For $n = 3$:
$$
x_3 = g(x_2) = \frac{\cos(0.595296) + 1}{3} = 0.609328
$$

For $n = 4$:
$$
x_4 = g(x_3) = \frac{\cos(0.609328) + 1}{3} = 0.6066778
$$

For $n = 5$:
$$
x_5 = g(x_4) = \frac{\cos(0.6066778) + 1}{3} = 0.607182
$$

For $n = 6$:
$$
x_6 = g(x_5) = \frac{\cos(0.607182) + 1}{3} = 0.607086
$$

For $n = 7$:
$$
x_7 = g(x_6) = \frac{\cos(0.607086) + 1}{3} = 0.607105
$$

For $n = 8$:
$$
x_8 = g(x_7) = \frac{\cos(0.607105) + 1}{3} = 0.607101 \approx 0.6071
$$

The approximate root of $\cos(x) = 3x - 1$ by the fixed-point iteration method is $0.6071$.

## Python Code Example

Below is an example of Python code that implements the Fixed Point Iteration Method.

```python
import numpy as np
import matplotlib.pyplot as plt

def fixed_point_iteration(g, x0, tol=1e-5, max_iter=100):
    """Fixed Point Iteration Method."""
    x = x0
    for i in range(max_iter):
        x_new = g(x)
        if abs(x_new - x) < tol:
            return x_new, i+1
        x = x_new
    raise ValueError("Did not converge within the maximum number of iterations")

# Define the function g(x) = (cos(x) + 1) / 3
g = lambda x: (np.cos(x) + 1) / 3

# Initial guess
x0 = 0

# Perform Fixed Point Iteration
root, iterations = fixed_point_iteration(g, x0)
print(f"Root: {root:.4f}, Iterations: {iterations}")

# Plotting the function and the fixed points
x = np.linspace(-1, 1, 400)
y = (np.cos(x) + 1) / 3

plt.plot(x, y, label='$g(x) = \\frac{\\cos(x) + 1}{3}$')
plt.plot(x, x, label='$y = x$', linestyle='--')
plt.scatter(root, g(root), color='red')
plt.title('Fixed Point Iteration')
plt.xlabel('$x$')
plt.ylabel('$g(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
